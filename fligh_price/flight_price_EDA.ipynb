```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
%matplotlib inline
```

```python
df = pd.read_excel("./flight_price.xlsx")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Date"] = df["Date_of_Journey"].str.split("/")
df["Date"].head()
```

```python
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
df.head()
```

```python
df.info()
```

```python
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
```

```python
df.info()
```

```python
df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df["Arrival_Time"].str.split(" ").str[0]
```

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(lambda x: x.split(" ")[0])
```

```python
df["Arrival_Hour"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_Mintue"] = df["Arrival_Time"].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df["Arrival_Hour"] = df["Arrival_Hour"].astype(int)
df["Arrival_Mintue"] = df["Arrival_Mintue"].astype(int)
```

```python
df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df["Depature_Hour"] = df["Dep_Time"].str.split(":").str[0]
df["Depature_Mintue"] = df["Dep_Time"].str.split(":").str[1]
```

```python
df["Depature_Hour"] = df["Depature_Hour"].astype(int)
df["Depature_Mintue"] = df["Depature_Mintue"].astype(int)
```

```python
df.drop("Dep_Time", axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df["Total_Stops"].unique()
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df["Total_Stops"].mode()
```

```python
df["Total_Stops"] = df["Total_Stops"].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 2, '4 stops': 4, np.nan: 1})
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df.drop("Route", axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df["Duration_Hours"] = df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
df["Duration_mintues"] = df["Duration"].str.split(" ").str[1].str.split("m").str[0]
df["Duration_mintues"] = df["Duration_mintues"].fillna(0)
```

```python
df.head(10)
```

```python
df.drop("Duration", axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Airline"].unique()
```

```python
encorder = OneHotEncoder()
```

```python
encorder.fit_transform(df[["Airline", "Source", "Destination"]]).toarray()
```

```python
encoded_df = pd.DataFrame(encorder.fit_transform(df[["Airline", "Source", "Destination"]]).toarray(), columns=encorder.get_feature_names_out())
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.drop(columns=["Airline", "Source", "Destination"], inplace=True)
```
